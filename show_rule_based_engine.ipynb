```python
import os
import json
from sklearn.metrics import ConfusionMatrixDisplay
from copyrightDet.rule_based import RuleBased
from copyrightDet.match_string import MatchString
import matplotlib.pyplot as plt

with open(os.path.join(os.getcwd(), "data", "pos_neg_copy_y.json"), "r", encoding="utf8") as file:
    data = json.load(file)

real_values = [float(i)*10 for i in data.values()]
corpus = data.keys()

rule_based = RuleBased()
result = [float(i)*10 for i in rule_based.predict(corpus)]

ConfusionMatrixDisplay.from_predictions(real_values, result, display_labels=[0, 0.5, 1])
```

```python
print("Documents, that wrongfully got classified as no Copyright Statement:")
for index in range(len(real_values)):
    if real_values[index] == 0.0 and result[index] == 10.0:
        print(list(corpus)[index])
            
    
```

```python
print("Documents, that wrongfully got classified as a Copyright Statement:")
for index in range(len(real_values)):
    if real_values[index] == 10.0 and result[index] == 0.0:
        print(list(corpus)[index])
```
